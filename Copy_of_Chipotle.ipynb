### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [URL](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
df.dtypes
df.describe
df.shape
```

### Step 4. Sort by the name of the item

```python
dfs=df.sort_values(by="item_name")
dfs.head(10)
```

### Step 5. What was the quantity of the most expensive item ordered?

```python
dfs_price = df.sort_values(by="quantity", ascending=False)
dfs_price["item_price"] = dfs_price["item_price"].str.replace(r'$',"").astype("float")
#print(dfs_price)
dfs_price.head(2)
```

### Step 6. How many times was a Veggie Salad Bowl ordered?

```python
df_salad = df[df["item_name"]=="Veggie Salad Bowl"]
len(df_salad)
```

### Step 7. How many times did someone order more than one Canned Soda?

```python
st7 = df[(df.item_name== "Canned Soda") & (df.quantity >1)]
len(st7)
#ans 20
```

### Step 8. How many products cost more than $10.00?

```python
if df.item_price[0][0] == '$':
  prices = [float(value[1:-1]) for value in df.item_price]
df.item_price = prices
print(df.query('item_price > 10').item_name.nunique())
df.head()
#ans 31
```
